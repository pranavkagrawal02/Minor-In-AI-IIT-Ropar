```python
# prompt: throw a dice 100 times. Plot a graph for frequency. Plot histogram.

import random
import matplotlib.pyplot as plt

# Throw a dice 100 times
rolls = [random.randint(1, 6) for _ in range(100)]

# Count the frequency of each outcome
frequencies = {}
for roll in rolls:
  frequencies[roll] = frequencies.get(roll, 0) + 1

# Plot the frequency
plt.figure(figsize=(8, 6))
plt.bar(frequencies.keys(), frequencies.values())
plt.xlabel("Dice Roll")
plt.ylabel("Frequency")
plt.title("Frequency of Dice Rolls (100 Rolls)")
plt.xticks(list(frequencies.keys()))
plt.show()


#Plot Histogram
plt.figure(figsize=(8, 6))
plt.hist(rolls, bins=range(1, 8), align='left', rwidth=0.8) #bins are 1-6
plt.xlabel("Dice Roll")
plt.ylabel("Frequency")
plt.title("Histogram of Dice Rolls (100 Rolls)")
plt.xticks(range(1, 7)) # Display ticks from 1-6
plt.show()
```

```python
# prompt: Generate me a example for Poisson distribution and plot a graph

import random
import matplotlib.pyplot as plt
import numpy as np

# Simulate Poisson distribution
# Example: Number of cars passing a certain point on a highway per hour (lambda = 10)
lambda_param = 10  # Average number of cars per hour
num_hours = 100  # Number of hours to simulate
cars_per_hour = np.random.poisson(lambda_param, num_hours)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(cars_per_hour, bins=range(min(cars_per_hour), max(cars_per_hour) + 2), align='left', rwidth=0.8)
plt.xlabel("Number of Cars")
plt.ylabel("Frequency")
plt.title(f"Poisson Distribution (Î» = {lambda_param}, {num_hours} hours)")
plt.xticks(range(min(cars_per_hour), max(cars_per_hour) + 1))  # Adjust x-ticks based on the data
plt.show()
```

```python
# prompt: Plot me a binomial distribution graph with 3 probabilities

import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10  # Number of trials
p_values = [0.2, 0.5, 0.8]  # Probabilities of success

# Create the plot
plt.figure(figsize=(10, 6))

for p in p_values:
    # Generate the binomial distribution
    x = range(n + 1)
    probabilities = binom.pmf(x, n, p)

    # Plot the distribution
    plt.plot(x, probabilities, marker='o', linestyle='-', label=f'p = {p}')

# Customize the plot
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distributions')
plt.xticks(x)  # Ensure all x-values are displayed
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: A new model has be designed to detect the face. This is compared over 10 data sets. New model and old model are compared over 10 data sets. Simulate the values. write null and alternate hypothesis. perform t test and present p value along with all the data and results. Plot necessary graphs

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for new and old models across 10 datasets
np.random.seed(42)  # for reproducibility
datasets = range(1, 11)
new_model_accuracy = np.random.uniform(0.85, 0.95, 10)
old_model_accuracy = np.random.uniform(0.75, 0.88, 10)

# Create a DataFrame for better visualization and analysis
data = pd.DataFrame({'Dataset': datasets,
                    'New Model': new_model_accuracy,
                    'Old Model': old_model_accuracy})

# Calculate the difference in accuracy
data['Difference'] = data['New Model'] - data['Old Model']


# Hypothesis Testing
# Null hypothesis (H0): There is no significant difference in accuracy between the new and old models.
# Alternative hypothesis (H1): There is a significant difference in accuracy between the new and old models.


# Perform a paired t-test
t_statistic, p_value = stats.ttest_rel(data['New Model'], data['Old Model'])


# Print results
print("Data:")
print(data)
print("\nHypothesis Test:")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a statistically significant difference between the new and old model accuracies.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude a significant difference.")

# Plotting

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data[['New Model', 'Old Model']])
plt.title('Accuracy Comparison: New vs. Old Model')
plt.ylabel('Accuracy')
plt.show()

# Scatter plot with difference
plt.figure(figsize=(8, 6))
plt.scatter(data['Dataset'], data['Difference'])
plt.axhline(y=0, color='red', linestyle='--') # Add a horizontal line at y=0 for visual reference
plt.title('Difference in Accuracy Between New and Old Models')
plt.xlabel('Dataset')
plt.ylabel('Accuracy Difference (New - Old)')
plt.show()
```
