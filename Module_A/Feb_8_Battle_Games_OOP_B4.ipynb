```python
class Warrior(Player):
    def use_special(self, opponent):
        if not self.specialUsed:
            self.specialUsed=True
            print("Warrior's Fury! Increased damage this round!")
            damage = random.randint(20, 30)
            print("Generating attack.....ğŸ’¥")
            time.sleep(2)
            if opponent.defending:
                damage = damage//2
                opponent.defending = False
            opponent.hp -= damage
            print(self.name, "attacks", opponent.name, "for", damage, "damage!âš”ï¸")
        else:
            print("Special ability already used!")

class Healer(Player):
    def use_special(self, opponent):
        if not self.specialUsed:
            self.specialUsed=True
            self.hp += 20
            if self.hp>100:
                self.hp = 100
            print(self.name, "heals themselves for 20 HP!âœ¨")
        else:
            print("Special ability already used!")

class Rogue(Player):
     def use_special(self, opponent):
        if not self.specialUsed:
             self.special_used = True
             print("Rogue Shot! Defense ignored!")
             damage = random.randint(10,20)
             opponent.hp = opponent.hp - damage
             opponent.defending = False
             print(self.name, "attacks", opponent.name, "for", damage, "damage!âš”ï¸")
        else:
            print("Special Ability has been used!")

```

```python
# @title
print("Game Starts")
player1_choice = input("Player 1: Choose your character (Warrior, Healer, Rogue): ")
player2_choice = input("Player 2: Choose your character (Warrior, Healer, Rogue): ")

characters = {
        "warrior": Warrior,
        "healer": Healer,
        "rogue": Rogue
    }

player1 = characters[player1_choice]("Player 1")
player2 = characters[player2_choice]("Player 2")

current_player = player1
opponent = player2

while player1.hp > 0 and player2.hp > 0:
        print("\n-------Turn Start---------")
        print(f"{player1.name}'s HP:", player1.hp)
        print(f"{player2.name}'s HP:", player2.hp)
        print(current_player.name, "'s turn")
        action = input('Choose attack, defend or special: ')

        if(action=='attack'):
            current_player.attack(opponent)
        elif(action == 'defend'):
            current_player.defend()
        elif(action == 'special'):
            current_player.use_special(opponent)

        current_player, opponent = opponent, current_player

winner = player1 if player1.hp > 0 else player2
print(f"{winner.name} wins the battle!ğŸ†")

```
