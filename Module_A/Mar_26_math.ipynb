```python
import numpy as np
import matplotlib.pyplot as plt

# Define a vector
v = np.array([3, 4])

# Create the plot
plt.figure(figsize=(5, 5))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid()

# Plot the vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')

# Set limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Visualization')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two vectors
A = np.array([3, 4])
B = np.array([4, -3])

# Compute the dot product
dot_product = np.dot(A, B)

# Create the plot
fig, ax = plt.subplots(figsize=(6, 6))
ax.axhline(0, color='gray', lw=1)
ax.axvline(0, color='gray', lw=1)
ax.grid()

# Plot the vectors
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')

# Show the dot product as text
ax.text(1, -4, f'Dot Product = {dot_product}', fontsize=12, color='black')

# Set limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

# Labels and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Dot Product Visualization')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a 3D vector
v = np.array([3, 4, 5])

# Create a 3D figure
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the vector
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='r', arrow_length_ratio=0.1)

# Set limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

# Labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Vector Visualization')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([3, 4])
v2 = np.array([1, 3])
sum_v = v1 + v2

plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid()

# Plot vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')
plt.quiver(0, 0, sum_v[0], sum_v[1], angles='xy', scale_units='xy', scale=1, color='g', label='sum_v')

# Labels
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Addition')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([3, 4])
magnitude = np.linalg.norm(v1)

plt.figure(figsize=(5, 5))
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.grid()

# Plot vector
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label=f'|v1| = {magnitude:.2f}')

# Set limits
plt.xlim(-1, 6)
plt.ylim(-1, 6)

# Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Magnitude of a Vector')
plt.legend()
plt.show()
```

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matrix Addition
addition = A + B
print("Matrix Addition:\n", addition)

# Matrix Multiplication
multiplication = np.dot(A, B)
print("\nMatrix Multiplication:\n", multiplication)

# Transpose of a Matrix
transpose_A = np.transpose(A)
print("\nTranspose of A:\n", transpose_A)

# Rank of a Matrix
rank_A = np.linalg.matrix_rank(A)
print("\nRank of A:", rank_A)

# Eigenvalues of a Matrix
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues of A:", eigenvalues)
print("\nEigenvectors of A:\n", eigenvectors)

# Determinant of a Matrix
determinant_A = np.linalg.det(A)
print("\nDeterminant of A:", determinant_A)
```

```python
import numpy as np

# 1. Create an array
arr = np.array([1, 2, 3, 4, 5])
print("1. Array Creation:\n", arr)

# 2. Reshape an array
matrix = np.arange(1, 10).reshape(3,3)
print("\n2. Reshaped Array (3x3 Matrix):\n", matrix)

# 3. Flattening an array
flattened = matrix.flatten()
print("\n3. Flattened Matrix:\n", flattened)

# 4. Element-wise operations
squared = arr ** 2
print("\n4. Element-wise Squaring:\n", squared)

# 5. Mean and standard deviation
mean_val = np.mean(arr)
std_dev = np.std(arr)
print("\n5. Mean:", mean_val, "Standard Deviation:", std_dev)

# 6. Generate random numbers
random_arr = np.random.rand(3, 3)
print("\n6. Random 3x3 Array:\n", random_arr)

# 7. Cumulative sum
cumsum_arr = np.cumsum(arr)
print("\n7. Cumulative Sum of Array:\n", cumsum_arr)

# 8. Identity matrix
identity_matrix = np.eye(3)
print("\n8. Identity Matrix:\n", identity_matrix)

# 9. Maximum and minimum values
max_value = np.max(arr)
min_value = np.min(arr)
print("\n9. Max Value:", max_value, "Min Value:", min_value)

# 10. Unique elements in an array
arr1 = [1, 2, 2, 1, 4, 5, 8, 6, 4]
unique_elements = np.unique(arr1)
print("\n10. Unique Elements in Array:\n", unique_elements)
```

```python
import pandas as pd

# 1. Create a DataFrame
data = {'Name': ['Akhil', 'Aryan', 'Aakash', 'Aya'], 'Age': [25, 22, 23, 24], 'City': ['Pune', 'Mumbai', 'Pune', 'Delhi']}
df = pd.DataFrame(data)
print("1. DataFrame:\n", df)

# 2. Display basic info
print("\n2. Basic Info:")
df.info()

# 3. Display summary statistics
print("\n3. Summary Statistics:\n", df.describe())

# 4. Select a single column
print("\n4. Selecting a Column (Age):\n", df['Age'])

# 5. Add a new column
df['Salary'] = [50000, 60000, 55000, 70000]
print("\n5. Adding a Column:\n", df)

# 6. Filter rows based on condition
filtered_df = df[df['Age'] > 23]
print("\n6. Filtering Rows (Age > 23):\n", filtered_df)

# 7. Sorting by a column
df_sorted = df.sort_values(by='Age')
print("\n7. Sorting by Age:\n", df_sorted)

# 8. Grouping data
grouped = df.groupby('City')['Salary'].mean()
print("\n8. Grouping by City (Mean Salary):\n", grouped)

# 9. Handling missing values
df.loc[1, 'Salary'] = None
print("\n9. DataFrame with Missing Value:\n", df)
df_filled = df.fillna(df['Salary'].mean())
print("\n9. Filling Missing Value with Mean:\n", df_filled)

# 10. Exporting to CSV
df.to_csv('data.csv')
print("\n10. Data exported to 'data.csv'")
```
