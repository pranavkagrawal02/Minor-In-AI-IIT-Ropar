```python
import numpy as np
import time

# Define transition matrix
transition_matrix = np.array([
    [0.7, 0.2, 0.1],  # Scrolling
    [0.5, 0.3, 0.2],  # Liking/Commenting
    [0.8, 0.15, 0.05] # Closing
])

# States
actions = ["Scrolling Reels", "Liking/Commenting", "Closing the App"]

# Start with scrolling
current_state = 0  # 0: Scrolling, 1: Liking/Commenting, 2: Closing

# Simulate for 15 minutes
for minute in range(1, 16):
    next_state = np.random.choice([0, 1, 2], p=transition_matrix[current_state])
    print(f"Minute {minute}: {actions[next_state]}")
    current_state = next_state
    time.sleep(5)  # Wait for 5 seconds before next iteration
```

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

# Define the states
states = ["Scrolling", "Liking/Commenting", "Closing App"]

# Define the transition matrix
transition_matrix = np.array([
    [0.7, 0.2, 0.1],  # Scrolling -> (Scrolling, Liking/Commenting, Closing)
    [0.5, 0.3, 0.2],  # Liking/Commenting -> (Scrolling, Liking/Commenting, Closing)
    [0.8, 0.15, 0.05] # Closing -> (Scrolling, Liking/Commenting, Closing)
])

# Create a directed graph
G = nx.DiGraph()

# Add nodes (states)
for state in states:
    G.add_node(state)

# Add edges (transitions) with probabilities as labels
for i in range(len(states)):
    for j in range(len(states)):
        if transition_matrix[i, j] > 0:
            G.add_edge(states[i], states[j], weight=transition_matrix[i, j])

# Positioning the nodes
pos = nx.spring_layout(G, seed=42)

# Draw the nodes and edges
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=3000, edge_color="gray", font_size=10, font_weight="bold")

# Add edge labels (transition probabilities)
edge_labels = {(states[i], states[j]): f"{transition_matrix[i, j]:.2f}" for i in range(len(states)) for j in range(len(states)) if transition_matrix[i, j] > 0}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

plt.title("Markov Chain: Instagram Reels Usage")
plt.show()
```

```python
import random

# Independent Events: Tossing a coin and rolling a die
def independent_events():
    coin_sample_space = ["Heads", "Tails"]
    dice_sample_space = [1, 2, 3, 4, 5, 6]

    coin_toss = random.choice(coin_sample_space)
    dice_roll = random.choice(dice_sample_space)

    print("Coin Toss Outcome:", coin_toss)
    print("Dice Roll Outcome:", dice_roll)

    # Multiplication Rule for Independent Events
    P_heads = 1/2
    P_six = 1/6
    P_heads_and_6 = P_heads * P_six
    print("P(Heads AND rolling a 6):", round(P_heads_and_6, 2))
    print("These events are independent.\n")

# Dependent Events: Picking two balls without replacement
def dependent_events():
    bag = ["Red", "Blue", "Green", "Yellow"]

    first_pick = random.choice(bag)
    bag.remove(first_pick)  # Removing the picked ball affects the next pick

    second_pick = random.choice(bag)

    print("First ball picked:", first_pick)
    print("Second ball picked:", second_pick)

    # Multiplication Rule for Dependent Events
    P_first_red = 1/4
    P_second_blue_given_red_picked = 1/3
    P_red_and_blue = P_first_red * P_second_blue_given_red_picked
    print("P(Red first AND Blue second):", round(P_red_and_blue, 2))
    print("These events are dependent (probability of second pick changes after the first).\n")


# Run all cases
independent_events()
dependent_events()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)
y1 = x ** 2
y2 = np.sin(x)

# Line Plot
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='y = x^2', marker='o', linestyle='-')
plt.plot(x, y2, label='y = sin(x)', marker='s', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.legend()
plt.grid()
plt.show()

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()

# Scatter Plot
x_scatter = np.random.rand(20) * 10
y_scatter = np.random.rand(20) * 10
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, color='orange', edgecolors='black')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot Example')
plt.grid()
plt.show()

# Histogram
data = np.random.randn(1000)  # Generating random data
plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set theme
sns.set_theme(style="darkgrid")

# Sample dataset
np.random.seed(42)
data = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'X': np.random.rand(100) * 10,
    'Y': np.random.rand(100) * 10,
    'Value': np.random.randn(100) * 5 + 50
})

# Line Plot
plt.figure(figsize=(6, 4))
sns.lineplot(x='X', y='Y', data=data)
plt.title("Line Plot Example")
plt.show()

# Bar Plot (Fixed)
plt.figure(figsize=(6, 4))
sns.barplot(x='Category', y='Value', hue='Category', data=data, palette='coolwarm', dodge=False)
plt.title("Bar Plot Example")
plt.show()

# Scatter Plot (Fixed: Better Visibility)
plt.figure(figsize=(6, 4))
sns.scatterplot(x='X', y='Y', hue='Category', data=data, palette='viridis', alpha=0.7, edgecolor="black")
plt.title("Scatter Plot Example")
plt.show()

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(data['Value'], bins=20, kde=True, color='blue')
plt.title("Histogram Example")
plt.show()

# Box Plot (Fixed)
plt.figure(figsize=(6, 4))
sns.boxplot(x='Category', y='Value', hue='Category', data=data, palette='Set2', dodge=True)
plt.title("Box Plot Example")
plt.show()

# Heatmap (Correlation Matrix)
plt.figure(figsize=(6, 4))
corr = data[['X', 'Y', 'Value']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap Example")
plt.show()
```
