```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Generate normally distributed data
data = np.random.normal(loc=50, scale=15, size=1000)  # Continuous values

# Compute KDE
kde = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)  # Generate x values for KDE
kde_values = kde(x_vals)  # Get density values

# Find the mode (highest density point)
mode_kde = x_vals[np.argmax(kde_values)]

# Plot histogram and KDE
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, stat="density", alpha=0.5, edgecolor="black", label="Histogram")
plt.plot(x_vals, kde_values, color='red', label="KDE Curve")

# Mark the mode
plt.axvline(mode_kde, color='blue', linestyle='dashed', linewidth=2, label=f"Mode (KDE)  {mode_kde:.2f}")

plt.legend()
plt.title("Finding Mode Using KDE")
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.show()

print(f"Estimated Mode (KDE): {mode_kde:.2f}")
```

```python
import scipy.stats as stats
import numpy as np

# Sample salary data
salaries_A = np.array([55, 60, 62, 58, 63, 59, 61, 60, 57, 64, 62, 56, 59, 58, 65])
salaries_B = np.array([52, 55, 57, 53, 58, 54, 56, 55, 51, 59, 57, 50, 54, 53, 60])

# Perform Levene's test
levene_stat, p_value = stats.levene(salaries_A, salaries_B)

print(f"Levene's Test Statistic: {levene_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05
if p_value > alpha:
    print("Variances are equal (p > 0.05), proceed with standard t-test.")
else:
    print("Variances are not equal (p  0.05), use Welch's t-test instead.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate x values (standard normal)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="Null Distribution (H True)")

# Shade rejection region (p < 0.05)
plt.fill_between(x, y, where=(x > 1.96), color='red', alpha=0.5, label="Reject H (p < 0.05)")
plt.fill_between(x, y, where=(x < -1.96), color='red', alpha=0.5)

# Critical values at 5% significance
plt.axvline(1.96, color='black', linestyle='dashed')
plt.axvline(-1.96, color='black', linestyle='dashed')

plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.title("Decision Making in Hypothesis Testing")
plt.legend()
plt.show()
```

```python
# Z Test - One Sample
import numpy as np
import scipy.stats as stats

# Given data
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=7.5, scale=1.5, size=40)  # Simulating student sleep data

population_mean = 8.0   # Claimed mean
population_std = 1.5    # Known population standard deviation
n = len(sample_data)    # Sample size

# Calculate sample mean
sample_mean = np.mean(sample_data)

# Compute Z-statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Compute p-value (one-tailed test)
p_value = stats.norm.cdf(z_score)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: Students sleep significantly less than 8 hours per night.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that students sleep less than 8 hours per night.")
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
population_mean = 8.0   # Claimed mean
population_std = 1.5    # Known population standard deviation
sample_size = 40        # Sample size

# Simulating sample data (students' sleep hours)
np.random.seed(42)
sample_data = np.random.normal(loc=7.5, scale=1.5, size=sample_size)

# Compute Z-score
sample_mean = np.mean(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value (left-tailed test)
p_value = stats.norm.cdf(z_score)

# Generate x values for normal distribution
x = np.linspace(-4, 4, 1000)  # Standard normal range
y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution (mean=0, std=1)

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade the left tail (p-value region)
x_fill = np.linspace(-4, z_score, 100)
ax.fill_between(x_fill, stats.norm.pdf(x_fill, 0, 1), color='red', alpha=0.4, label="P-value region")

# Mark the Z-score
ax.axvline(z_score, color='red', linestyle='dashed', label=f"Z-Score = {z_score:.2f}")

# Mark the critical region ( = 0.05)
critical_z = stats.norm.ppf(0.05)  # Critical Z-score for 5% significance (left-tailed test)
ax.axvline(critical_z, color='black', linestyle='dotted', label=f"Critical Z = {critical_z:.2f}")

# Display p-value on the graph
ax.text(z_score, 0.05, f"P-value = {p_value:.4f}", color='red', fontsize=10, ha='left', fontweight='bold')

# Labels and title
ax.set_title("One-Sample Z-Test Visualization", fontsize=12)
ax.set_xlabel("Z-Score", fontsize=10)
ax.set_ylabel("Probability Density", fontsize=10)
ax.legend()

# Show plot
plt.show()
```

```python
# Checking batter life
# Null: The average battery life is 10 hours
# Alternate: The average battery life is not 10 hours

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
population_mean = 10.0   # Claimed mean
population_std = 1.5    # Known population standard deviation
sample_size = 50        # Sample size

# Simulating sample data (actual measured battery life)
np.random.seed(42)
sample_data = np.random.normal(loc=9.6, scale=1.5, size=sample_size)

# Compute sample mean
sample_mean = np.mean(sample_data)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Compute p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Multiply by 2 for two-tailed test

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
```

```python
# Define significance level
alpha = 0.05

# Decision rule
if p_value < alpha:
    print("Reject the null hypothesis (H0). The battery life is significantly different from 10 hours.")
else:
    print("Fail to reject the null hypothesis (H0). No significant difference in battery life.")
```

```python
# Chi Square : Association between two catogorical values
# Null: Gender and mobile brand preference are independent (no relationship)
# Alternate: Gender and mobile brand preference are dependent (related)

import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Observed data (contingency table)
observed = np.array([[30, 40],  # Apple
                     [50, 30],  # Samsung
                     [20, 30]]) # OnePlus

# Perform Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

# Print results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Decision Rule
alpha = 0.05
if p_value < alpha:
    print("Reject the Null Hypothesis (There is a relationship).")
else:
    print("Fail to Reject the Null Hypothesis (No significant relationship).")
```

```python
# Convert data to DataFrame
brands = ["Apple", "Samsung", "OnePlus"]
gender = ["Male", "Female"]
observed_df = pd.DataFrame(observed, index=brands, columns=gender)
expected_df = pd.DataFrame(expected, index=brands, columns=gender)

# Plot heatmap for observed and expected values
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(observed_df, annot=True, cmap="Blues", fmt=".0f", ax=axes[0])
axes[0].set_title("Observed Frequencies")

sns.heatmap(expected_df, annot=True, cmap="Reds", fmt=".2f", ax=axes[1])
axes[1].set_title("Expected Frequencies")

plt.show()
```

```python
from graphviz import Digraph
from IPython.display import display, Image

def create_decision_flowchart():
    dot = Digraph()

    # Nodes
    dot.node('Start', 'Start')
    dot.node('Q1', 'Is your data Numerical?')
    dot.node('Q2', 'Are you comparing means?')
    dot.node('Q3', 'Comparing One-Sample vs Population Mean?')
    dot.node('Q4', 'Population Variance Known?')
    dot.node('Q5', 'Are the samples paired?')
    dot.node('Q6', 'Population Variance Known?')
    dot.node('Q7', 'Do samples have equal variance?')

    # Decisions
    dot.node('ChiSquare', 'Use Chi-Square Test', shape='box')
    dot.node('NotApplicable', 'Not a T-Test or Z-Test', shape='box')
    dot.node('ZTest1', 'Use Z-Test', shape='box')
    dot.node('OneTTest', 'Use One-Sample T-Test', shape='box')
    dot.node('PairedTTest', 'Use Paired T-Test', shape='box')
    dot.node('ZTest2', 'Use Two-Sample Z-Test', shape='box')
    dot.node('TTest', 'Use Standard Two-Sample T-Test', shape='box')
    dot.node('WelchTTest', 'Use Welchs T-Test', shape='box')

    # Edges
    dot.edge('Start', 'Q1')
    dot.edge('Q1', 'ChiSquare', label='No')
    dot.edge('Q1', 'Q2', label='Yes')
    dot.edge('Q2', 'NotApplicable', label='No')
    dot.edge('Q2', 'Q3', label='Yes')
    dot.edge('Q3', 'Q4', label='Yes')
    dot.edge('Q4', 'ZTest1', label='Yes')
    dot.edge('Q4', 'OneTTest', label='No')
    dot.edge('Q3', 'Q5', label='No')
    dot.edge('Q5', 'PairedTTest', label='Yes')
    dot.edge('Q5', 'Q6', label='No')
    dot.edge('Q6', 'ZTest2', label='Yes')
    dot.edge('Q6', 'Q7', label='No')
    dot.edge('Q7', 'TTest', label='Yes')
    dot.edge('Q7', 'WelchTTest', label='No')

    return dot

# Generate the flowchart and display it in Colab
flowchart = create_decision_flowchart()
flowchart.format = 'png'
flowchart.render('decision_flowchart', format='png', cleanup=False)
display(Image(filename='decision_flowchart.png'))
```
