# Matplotlib

What is Matplotlib and Why do we use it ?
Matplotlib is a 2D plotting library that allows you to create a wide range of visualizations, such as line plots, bar charts, histograms, scatter plots, and more.
- Data Insights
- Customization
- Integration
- Versatility

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

- figures : The entire canvas where plots are drawn
- Axes : The individual plot area
- Elements : Lines, bars, points , titles , legends , etc.

create figure -> plot data on the axes -> Customize -> Display the plot.

```python

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)  # Line plot
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Zepto Sales Anually since 2021")
plt.show()  # Display the plot
```

## Visualazing Distributions

```python
#Bar chart compares sales across products
products = ['A', 'B', 'C']
sales = [100, 150, 80]

plt.bar(products, sales, color=['red', 'green', 'blue'])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales by Product")
plt.show()

#Histogram shows the spread of random data
data = np.random.randn(1000)  # 1000 random numbers
plt.hist(data, bins=30, color='purple', alpha=0.1)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Distribution of Random Data")
plt.show()
```

## Comparing Data

```python
#Scatter plots are ideal for showing relationships between two variables
height = [150, 160, 170, 180, 190]
weight = [50, 60, 65, 80, 90]

plt.scatter(height, weight, color='green', s=100, alpha=0.5)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs. Weight")
plt.show()
```

## Working with Pandas DF

```python
import pandas as pd

# Sample DataFrame
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
        'Revenue': [200, 220, 250, 300]}
df = pd.DataFrame(data)

# Plot
plt.plot(df['Month'], df['Revenue'], marker='o', color='orange')
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthly Revenue Trend")
plt.show()
```

## Customizing Plots

```python
# Multiple lines with legend
x = np.array([1, 2, 3, 4])
y1 = x**2
y2 = x**3

plt.plot(x, y1, label="Square", color='blue', marker='o')
plt.plot(x, y2, label="Cube", color='red', marker='s')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Square vs. Cube")
plt.legend()
plt.grid(True)
plt.show()
plt.savefig('plot.png')
```

# Lets do some analysis

A small retail business wants to understand its sales performance over the past year to:

1. Identify months with peak sales (e.g., for inventory planning).
2. Spot any declining trends (e.g., to  adjust marketing strategies).
3. Visualize the data to present to stakeholders.

## We create a dummy real world dataset

```python

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [20000, 22000, 25000, 23000, 21000, 24000, 26000, 27000, 25000, 28000, 32000, 35000],
    'Expenses': [15000, 16000, 17000, 16500, 15500, 18000, 19000, 20000, 18500, 19500, 21000, 22000]
}


df = pd.DataFrame(data)
print(df)
```

## Data Analysis of our data

```python
# we can think that we can find profit from our data set
df['Profit'] = df['Sales'] - df['Expenses']
print(df)
```

What can we observe ?
1. Profit fluctuates, with a noticeable increase toward the end of the year
2. May has the lowest profit (5500), while Dec has the highest (13,000)

## Visualization

lets plot our all the three features of our data using a line plot, and customize it for a good visualization

```python
# Line plot
plt.figure(figsize=(10, 6))  # Set figure size
plt.plot(df['Month'], df['Sales'], label='Sales', marker='o', color='blue',linestyle='--')
plt.plot(df['Month'], df['Expenses'], label='Expenses', marker='s', color='red')
plt.plot(df['Month'], df['Profit'], label='Profit', marker='^', color='green')

# Customization
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Monthly Sales, Expenses, and Profit (2024)')
plt.legend()  # Add legend
plt.grid(True, linestyle='--', alpha=0.7)  # Add grid
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

What can you obserserve from this ?

- Sales and Expenses rise steadily, but Sales grow faster toward year-end.
- Profit spikes in Nov-Dec, indicating a strong holiday season.
- Expenses do rise but has lower slope compared to profit and sales.

### Lets try to get some more insights

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['Profit'], color='teal', alpha=0.8)

plt.xlabel('Month')
plt.ylabel('Profit ($)')
plt.title('Monthly Profit (2024)')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Grid only on y-axis
plt.tight_layout()
plt.show()
```

we didn't have any signigicant observation but can say highest profit is in dec, whereas lowest is in may.

## Insights

1. Peak Sales Months: Nov and Dec show the highest sales and profit, likely due to holiday shopping. The business should stock up inventory and ramp up marketing in Q4.
2. Low Performer: May has the lowest profit (5500). Investigate whyperhaps low demand or high expensesand consider promotions or cost-cutting.
3. Trend: Sales outpace expenses toward year-end, improving profitability. This suggests a seasonal business cycle.
