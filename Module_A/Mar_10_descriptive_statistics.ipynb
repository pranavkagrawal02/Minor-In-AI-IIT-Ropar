```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import gaussian_kde

# Generate symmetric data (normal distribution)
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std Dev=10

# Compute mean, median, and mode
mean = np.mean(data)
median = np.median(data)

# Estimate mode using density function
density = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)  # Generate points for estimation
mode = x_vals[np.argmax(density(x_vals))]  # Find x-value with max density

# Create histogram
sns.histplot(data, bins=30, kde=True, color="royalblue")

# Plot mean, median, and mode
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='orange', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')

# Labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.legend()

# Show the plot
plt.show()

# Print values
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import gaussian_kde

# Generate right-skewed data (Exponential Distribution)
data = np.random.exponential(scale=10, size=1000)  # Skewed to the right

# Compute mean, median, and mode
mean = np.mean(data)
median = np.median(data)

# Estimate mode using Kernel Density Estimation (KDE)
density = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 1000)  # Generate points for estimation
mode = x_vals[np.argmax(density(x_vals))]  # Find x-value with max density

# Create histogram
sns.histplot(data, bins=30, kde=True, color="royalblue")

# Plot mean, median, and mode
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='orange', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')

# Labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram:")
plt.legend()

# Show the plot
plt.show()

# Print values
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Generate synthetic data (Normal Distribution with slight variation)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std Dev=15

# Compute Descriptive Statistics
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True)[0][0]  # Mode using scipy.stats
data_range = np.max(data) - np.min(data)
variance = np.var(data, ddof=1)  # Sample variance
std_dev = np.sqrt(variance)  # Standard deviation
q1 = np.percentile(data, 25)  # 25th percentile (Q1)
q3 = np.percentile(data, 75)  # 75th percentile (Q3)
iqr = q3 - q1  # Interquartile Range (Q3 - Q1)
percentile_90 = np.percentile(data, 90)  # 90th percentile

# Print Descriptive Statistics
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Range: {data_range:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
print(f"90th Percentile: {percentile_90:.2f}")

# Plot Histogram with KDE
sns.histplot(data, bins=30, kde=True, color="royalblue")

# Add vertical lines for mean, median, and mode
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.axvline(mode, color='orange', linestyle='dashed', linewidth=2, label=f'Mode: {mode:.2f}')
plt.axvline(q1, color='purple', linestyle='dashed', linewidth=2, label=f'Q1 (25%): {q1:.2f}')
plt.axvline(q3, color='brown', linestyle='dashed', linewidth=2, label=f'Q3 (75%): {q3:.2f}')
plt.axvline(percentile_90, color='cyan', linestyle='dashed', linewidth=2, label=f'90th Percentile: {percentile_90:.2f}')

# Labels and Title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Descriptive Statistics Visualization")
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate normally distributed data (perfectly symmetrical)
data = np.random.normal(loc=50, scale=10, size=500)  # Mean=50, Std Dev=10

# Compute quartiles and min/max
Q1 = np.percentile(data, 25)  # 25th percentile
Q2 = np.percentile(data, 50)  # Median (50th percentile)
Q3 = np.percentile(data, 75)  # 75th percentile
minimum = np.min(data)  # Minimum value
maximum = np.max(data)  # Maximum value

# Print all quartile values
print(f"Minimum: {minimum:.2f}")
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Median (Q2, 50th percentile): {Q2:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"Maximum: {maximum:.2f}")

# Create a DataFrame for visualization
df = pd.DataFrame({"Values": data, "Category": "Symmetric Data"})  # Single category for boxplot

# Create a perfect box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x="Category", y="Values", hue="Category", data=df, width=0.3,
            palette={"Symmetric Data": "royalblue"}, legend=False)

# Labels and title
plt.xlabel("Data Type")
plt.ylabel("Values")
plt.title("Perfectly Symmetric Box Plot")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic student scores (out of 100)
np.random.seed(42)
student_scores = np.random.normal(loc=70, scale=15, size=100)  # Mean=70, Std Dev=15
student_scores = np.clip(student_scores, 0, 100)  # Ensuring scores remain between 0-100

# Compute quartiles and min/max
Q1 = np.percentile(student_scores, 25)  # 25th percentile
Q2 = np.percentile(student_scores, 50)  # Median (50th percentile)
Q3 = np.percentile(student_scores, 75)  # 75th percentile
minimum = np.min(student_scores)  # Minimum score
maximum = np.max(student_scores)  # Maximum score

# Create a DataFrame for visualization
df = pd.DataFrame({"Scores": student_scores, "Category": "Student Scores"})

# Create the box plot
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x="Category", y="Scores", hue="Category", data=df, width=0.3,
                 palette={"Student Scores": "royalblue"}, legend=False)

# Add text annotations for min, Q1, median, Q3, and max
stat_values = [minimum, Q1, Q2, Q3, maximum]
stat_labels = ["Min", "Q1", "Median", "Q3", "Max"]

for i, val in enumerate(stat_values):
    plt.text(0, val, f"{stat_labels[i]}: {val:.2f}",
             horizontalalignment='right', verticalalignment='bottom',
             fontsize=10, fontweight='bold', color="black",
             bbox=dict(facecolor='white', alpha=0.6, edgecolor='black'))

# Labels and title
plt.xlabel("Data Type")
plt.ylabel("Scores (out of 100)")
plt.title("Student Scores Distribution (Box Plot)")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)

# Large IQR: Data is widely spread in the middle
data_large_iqr = np.concatenate([
    np.random.normal(loc=50, scale=5, size=50),  # Lower values (Q1)
    np.random.normal(loc=75, scale=15, size=50)  # Higher values (Q3)
])

# Small IQR: Data is closely packed in the middle
data_small_iqr = np.random.normal(loc=60, scale=5, size=100)  # More tightly packed

# Combine into DataFrame
df = pd.DataFrame({
    "Values": np.concatenate([data_large_iqr, data_small_iqr]),
    "Category": ["Large IQR"] * len(data_large_iqr) + ["Small IQR"] * len(data_small_iqr)
})

# Create box plot
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x="Category", y="Values", hue="Category", data=df, width=0.4,
                 palette={"Large IQR": "royalblue", "Small IQR": "orange"}, legend=False)

# Compute and annotate quartiles
for category, color in [("Large IQR", "black"), ("Small IQR", "darkred")]:
    subset = df[df["Category"] == category]["Values"]
    Q1, Q2, Q3 = np.percentile(subset, [25, 50, 75])
    min_val, max_val = np.min(subset), np.max(subset)

    # Text annotations
    stat_values = [min_val, Q1, Q2, Q3, max_val]
    stat_labels = ["Min", "Q1", "Median", "Q3", "Max"]

    for val, label in zip(stat_values, stat_labels):
        plt.text(["Large IQR", "Small IQR"].index(category), val, f"{label}: {val:.2f}",
                 horizontalalignment='right', verticalalignment='bottom',
                 fontsize=10, fontweight='bold', color=color,
                 bbox=dict(facecolor='white', alpha=0.6, edgecolor=color))

# Labels and title
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Comparison: Large IQR vs Small IQR")

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate political survey data (bimodal distribution: extremes at -10 and 10)
np.random.seed(42)
extreme_left = np.random.normal(-9, 1, 150)  # Strongly left-leaning
extreme_right = np.random.normal(9, 1, 150)  # Strongly right-leaning
moderates = np.random.normal(0, 2, 100)  # Moderate views

# Combine into a single dataset
political_views = np.concatenate([extreme_left, extreme_right, moderates])

# Calculate quartiles
Q1 = np.percentile(political_views, 25)
median = np.percentile(political_views, 50)
Q3 = np.percentile(political_views, 75)

# Create figure with two subplots: Box Plot & Histogram
fig, axes = plt.subplots(2, 1, figsize=(8, 10))

# Box plot
sns.boxplot(y=political_views, color="royalblue", width=0.5, ax=axes[0])
axes[0].set_ylabel("Political View Scale (-10 = Extreme Left, 10 = Extreme Right)")
axes[0].set_title("Box Plot of Political Views in a Survey")

# Annotate quartiles on box plot
axes[0].axhline(Q1, color='orange', linestyle='dashed', label=f'Q1: {Q1:.2f}')
axes[0].axhline(median, color='red', linestyle='dashed', label=f'Median: {median:.2f}')
axes[0].axhline(Q3, color='green', linestyle='dashed', label=f'Q3: {Q3:.2f}')
axes[0].legend()

# Histogram
sns.histplot(political_views, bins=30, kde=True, color="royalblue", ax=axes[1])
axes[1].set_xlabel("Political View Scale (-10 to 10)")
axes[1].set_ylabel("Frequency")
axes[1].set_title("Histogram of Political Views in a Survey")

# Show the plots
plt.tight_layout()
plt.show()

# Print quartile values
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Median (50th percentile): {median:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate symmetric test scores (normally distributed around 75)
np.random.seed(42)
test_scores = np.random.normal(loc=75, scale=10, size=100)  # Mean 75, Std Dev 10

# Define outlier sets
few_outliers = np.append(test_scores, [120, 130])  # 2 outliers
moderate_outliers = np.append(test_scores, [120, 130, 135, 140])  # 4 outliers
many_outliers = np.append(test_scores, [134, 140, 120, 143, 123, 134])  # 6 outliers

# Compute means
mean_few = np.mean(few_outliers)
mean_moderate = np.mean(moderate_outliers)
mean_many = np.mean(many_outliers)

# Create box plots
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.boxplot(few_outliers, patch_artist=True)
plt.title("2 Outliers")
plt.ylabel("Test Scores")
plt.ylim(20, 150)
plt.xticks([1], ["Few Outliers"])
plt.text(1.1, mean_few, f"Mean: {mean_few:.1f}", color="red", fontsize=10)

plt.subplot(1, 3, 2)
plt.boxplot(moderate_outliers, patch_artist=True)
plt.title("4 Outliers")
plt.ylim(20, 150)
plt.xticks([1], ["Moderate Outliers"])
plt.text(1.1, mean_moderate, f"Mean: {mean_moderate:.1f}", color="red", fontsize=10)

plt.subplot(1, 3, 3)
plt.boxplot(many_outliers, patch_artist=True)
plt.title("6 Outliers")
plt.ylim(20, 150)
plt.xticks([1], ["Many Outliers"])
plt.text(1.1, mean_many, f"Mean: {mean_many:.1f}", color="red", fontsize=10)

plt.suptitle("Effect of Outliers on Mean")
plt.tight_layout()
plt.show()
```
